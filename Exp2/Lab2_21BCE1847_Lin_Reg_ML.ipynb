# Training a Linear Regression Model

```python
X = Star_bucks[[' TotalFat','TransFat','SaturatedFat',' Sodium']]
```

```python
Y=Star_bucks['Calories']
```

```python
X
```

```python
Y
```

# Split the data into training / testing set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=101)
```

# Building the Model

```python
from sklearn.linear_model import LinearRegression
```

```python
mylm = LinearRegression()
```

```python
mylm.fit(X_train,y_train)
```

# Model Evaluation

```python
print(mylm.intercept_)
```

```python
coeff_df = pd.DataFrame(mylm.coef_,X.columns,columns=['Coefficient'])
coeff_df
```

```python
mylm.coef_
```

# Predictions from our Model

```python
predictions = mylm.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
```

# Plot a histogram of the residuals and make sure it looks normally distributed

```python
sns.distplot((y_test-predictions),bins=50);
```

# Regression Evaluation Metrics
Mean Absolute Error (MAE) is the easiest to understand, because it's the average error.

Mean Squared Error (MSE) is more popular than MAE, because MSE "punishes" larger errors, which tends to be useful in the real world.

Root Mean Squared Error (RMSE) is even more popular than MSE, because RMSE is interpretable in the "y" units.

All of these are loss functions, because we want to minimize them.

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
