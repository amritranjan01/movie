Scaling the data

```python
# scaling the data
scaler=StandardScaler()
data_scaled=pd.DataFrame(scaler.fit_transform(data), columns=data.columns)
```

```python
data_scaled.head()
```

Principal Component Analysis¶
Apply PCA algorithm with number of components equal to the total number of columns in the data

```python
#Defining the number of principal components to generate 
n=data_scaled.shape[1]

#Finding principal components for the data
pca = PCA(n_components=n, random_state=1) #Apply the PCA algorithm with random state = 1

data_pca1 = pd.DataFrame(pca.fit_transform(data_scaled)) #Fit and transform the pca function on scaled data

#The percentage of variance explained by each principal component
exp_var = pca.explained_variance_ratio_
```

```python
# visulaize the explained variance by individual components
plt.figure(figsize = (10,10))
plt.plot(range(1,8), exp_var.cumsum(), marker = 'o', linestyle = '--')
plt.title("Explained Variances by Components")
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
```

```python
# find the least number of components that can explain more than 90% variance
sum = 0
for ix, i in enumerate(exp_var):
  sum = sum + i
  if(sum>0.90):
    print("Number of PCs that explain at least 90% variance: ", ix+1)
    break
```

```python
pc_comps = ['PC1','PC2','PC3']
data_pca = pd.DataFrame(np.round(pca.components_[:3,:],2),index=pc_comps,columns=data_scaled.columns)
data_pca.T
```

Visualize the data in 2 dimensions using first two principal components

```python
plt.figure(figsize = (7,7))
sns.scatterplot(x=data_pca1[0],y=data_pca1[1])
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

Let's try adding hue to the scatter plot

```python
df_concat = pd.concat([data_pca1, data], axis=1)

plt.figure(figsize = (7,7))
#Create a scatter plot with x=0 and y=1 using df_concat dataframe
sns.scatterplot(data = df_concat, x=df_concat[0], y=df_concat[1], hue ='cylinders')
plt.xlabel("PC1")
plt.ylabel("PC2")
```

t-SNE¶
Apply TSNE embedding with 2 components for the dataframe data_scaled

```python
#Apply the TSNE algorithm with random state = 1
tsne = TSNE(n_components = 2, random_state=1)  

#Fit and transform tsne function on the scaled data
data_tsne = tsne.fit_transform(data_scaled)
```

```python
data_tsne.shape
```

```python
data_tsne = pd.DataFrame(data = data_tsne, columns = ['Component 1', 'Component 2'])
```

```python
data_tsne.head()
```

```python
sns.scatterplot(x=data_tsne.iloc[:,0],y=data_tsne.iloc[:,1])
```

```python
# Lets see scatter plot of the data w.r.t number of cyinders
sns.scatterplot(x=data_tsne.iloc[:,0],y=data_tsne.iloc[:,1],hue=data.cylinders)
```

```python
# Lets assign points to the 3 different groups
def grouping(x):
    first_component = x['Component 1']
    second_component = x['Component 2']
    if (first_component> 0) and (second_component >0): 
        return 'group_1'
    if (first_component >-20 ) and (first_component < 5):
        return 'group_2'
    else: 
        return 'group_3'
```

```python
data_tsne['groups'] = data_tsne.apply(grouping,axis=1)
```

```python
sns.scatterplot(x=data_tsne.iloc[:,0],y=data_tsne.iloc[:,1],hue=data_tsne.iloc[:,2])
```

```python
data['groups'] = data_tsne['groups'] 
```

```python
all_col = data.columns.tolist()
plt.figure(figsize=(20, 20))

for i, variable in enumerate(all_col):
    if i==7:
        break
    plt.subplot(4, 2, i + 1)
    #Create boxplot with groups on the x-axis
    sns.boxplot(y=data[variable], x=data['groups'])
    plt.tight_layout()
    plt.title(variable)
plt.show()
```

Conclusion:

Principal Component Analysis (PCA) and t-distributed Stochastic Neighbor Embedding (t-SNE) are both powerful techniques used for dimensionality reduction and visualization of high-dimensional data.

After applying PCA to our dataset, we observed that the variance in the data could be explained by a relatively small number of principal components. By retaining the top k components, we were able to reduce the dimensionality of the dataset while preserving most of its variance. The visualization of the data in the reduced-dimensional space revealed clusters and patterns that suggest underlying structures or relationships within the data.

Subsequently, t-SNE was applied to further explore the structure of the data. Unlike PCA, t-SNE is a nonlinear technique that aims to preserve local structure, making it particularly useful for visualizing complex datasets with nonlinear relationships. The t-SNE visualization revealed distinct clusters and groupings that were not apparent in the PCA plot. This suggests that t-SNE was able to capture more intricate patterns in the data that may be missed by linear techniques like PCA.

Overall, the combination of PCA and t-SNE provided valuable insights into the structure and relationships within the dataset. While PCA offered a high-level overview and dimensionality reduction, t-SNE delved deeper into the local structure, uncovering finer details and clusters. These findings can guide further analysis and interpretation of the data, potentially leading to a better understanding of underlying patterns and phenomena.
