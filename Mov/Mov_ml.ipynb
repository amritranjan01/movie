```python
# Load the Excel file
excel_file = pd.ExcelFile('Multiclass_dataset.xlsx')
# Read each sheet into separate DataFrames
sheet1_df = excel_file.parse('trainingset')
sheet2_df = excel_file.parse('TestingSet')
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Load the data
data = pd.read_excel('Multiclass_dataset.xlsx')

# Preprocessing
features = data.drop('Target Label', axis=1)
labels = data['Target Label']

# Clustering techniques
clustering_algorithms = {
    'KMeans': KMeans(n_clusters=3),
    'Agglomerative': AgglomerativeClustering(n_clusters=3),
    'DBSCAN': DBSCAN(eps=0.3)
}

# Apply each technique
for name, algorithm in clustering_algorithms.items():
    # Clustering
    clusters = algorithm.fit_predict(features)
    
    # Performance analysis
    score = silhouette_score(features, clusters)
    print(f'Silhouette Score of {name}: {score}')
    
    # Visualization
    plt.figure(figsize=(10, 7))
    plt.scatter(features['P1'], features['P2'], c=clusters, cmap='viridis')
    plt.title(f'Clusters from {name}')
    plt.show()
```

```python
dataWithoutLabels = sheet1_df.drop(["Target Label"], axis = 1)
dataWithoutLabels.head()
```

```python
from sklearn.cluster import KMeans
wcss = [] # within cluster sum of squares

for k in range(1, 15):
    kmeansForLoop = KMeans(n_clusters = k)
    kmeansForLoop.fit(dataWithoutLabels)
    wcss.append(kmeansForLoop.inertia_)

plt.figure(figsize = (10, 10))
plt.plot(range(1, 15), wcss)
plt.xlabel("K value")
plt.ylabel("WCSS")
plt.show()
```

```python
# Elbow point starting from 2 

dataWithoutLabels = sheet1_df.loc[:,['P8','P9']]
kmeans = KMeans(n_clusters = 2)
clusters = kmeans.fit_predict(dataWithoutLabels)
dataWithoutLabels["type"] = clusters
dataWithoutLabels["type"].unique()
```

```python
plt.figure(figsize = (15, 10))
plt.scatter(dataWithoutLabels["P8"][dataWithoutLabels["type"] == 0], dataWithoutLabels["P9"][dataWithoutLabels["type"] == 0], color = "red")
plt.scatter(dataWithoutLabels["P8"][dataWithoutLabels["type"] == 1], dataWithoutLabels["P9"][dataWithoutLabels["type"] == 1], color = "green")
plt.xlabel('P8')
plt.ylabel('P9')
plt.show()
```

```python
dataWithoutDiagnosis = sheet1_df.drop(["Target Label"], axis = 1)
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
scalar = StandardScaler()
kmeans = KMeans(n_clusters = 2)
pipe = make_pipeline(scalar, kmeans)
pipe.fit(dataWithoutDiagnosis)
labels = pipe.predict(dataWithoutDiagnosis)
df = pd.DataFrame({'labels': labels, "Target Label" : sheet1_df['Target Label']})
ct = pd.crosstab(df['labels'], sheet1_df['Target Label'])
print(ct)
```
